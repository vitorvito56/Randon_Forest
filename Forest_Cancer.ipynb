```python
#This is  a simple random forest model to predict breast cancer, usinf the sklearn dataset.
# The data is already processed in a perfect way, so I dont gonna pre process it.

import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
data= load_breast_cancer()
cancer_data= pd.DataFrame(data=data['data'],columns=data['feature_names'])
cancer_data

# In this part, I load the dataset, and transform it so i can get a better vizualisation.
```

```python
tumor_data = data.target
tumor_data

#This is the data we want to predict if 0 being a Benign Tumor, that means you have to treat it\
# and are sick, but isnt cancer
# 1 means the worst cenario, that the tumor spread to the body and you have cancer
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cancer_data,tumor_data,test_size=0.3)

# In this part, I separated the train and test Data, with 30% being used to test.
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier()

rf_classifier.fit(x_train, y_train)

#This part we import the Randon Forest Model and Implement it.
```

```python
y_pred = rf_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
accuracy= accuracy_score(y_test,y_pred)
precision = precision_score(y_test, y_pred)
print(f" accuracy= {accuracy}, precision ={precision}")

#This is the evaluation of the model, that was really high
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer_data,tumor_data,test_size=0.3)

rf_classifier = RandomForestClassifier(bootstrap= False)

rf_classifier.fit(x_train, y_train)

y_pred = rf_classifier.predict(x_test)
accuracy= accuracy_score(y_test,y_pred)
precision = precision_score(y_test, y_pred)
print(f" accuracy= {accuracy}, precision ={precision}")

#In this part we do the same thing for before, doing the model, with one difference:
# We have bootstrap false, so instead of it using random samples in each tree, they gonna use the full data
# This increase the chances of overfitting and didnt change a lot about the results
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer_data,tumor_data,test_size=0.3)

rf_classifier = RandomForestClassifier(criterion= 'entropy')

rf_classifier.fit(x_train, y_train)

y_pred = rf_classifier.predict(x_test)
accuracy= accuracy_score(y_test,y_pred)
precision = precision_score(y_test, y_pred)
print(f" accuracy= {accuracy}, precision ={precision}")


# In this case i just decides to use the criterion, and the entropy did one almost perfect system.
# I couldnt ask for a better prediction.
```
